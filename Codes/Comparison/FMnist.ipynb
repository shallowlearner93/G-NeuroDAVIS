# Importing libraries

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)

import sys
sys.path.append(path)

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import ipynbname
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utils import ComparisonPlot as CP

dataname = ipynbname.name().rsplit("_")[0]
```

# Load results

```python
GDAVIS_knn = pd.read_csv(path_tab+dataname+'_NeuroGDAVIS_Unsupervised_knn.csv',index_col=0)
VAE_knn = pd.read_csv(path_tab+dataname+'_VAE_knn.csv',index_col=0)

GDAVIS_rfc = pd.read_csv(path_tab+dataname+'_NeuroGDAVIS_Unsupervised_rfc.csv',index_col=0)
VAE_rfc = pd.read_csv(path_tab+dataname+'_VAE_rfc.csv',index_col=0)

GDAVIS_trust = pd.read_csv(path_tab+dataname+'_NeuroGDAVIS_Unsupervised_trust.csv',index_col=0)
VAE_trust = pd.read_csv(path_tab+dataname+'_VAE_trust.csv',index_col=0)
```

```python
knn_df = {}
for i in GDAVIS_knn.columns:
    knn_df[i] = pd.DataFrame()
    knn_df[i]['G-NeuroDAVIS'] = GDAVIS_knn.loc[:,i]
    knn_df[i]['VAE'] = VAE_knn.loc[:,i]
    knn_df[i] = knn_df[i].T
    
rfc_df = {}
for i in GDAVIS_rfc.columns:
    rfc_df[i] = pd.DataFrame()
    rfc_df[i]['G-NeuroDAVIS'] = GDAVIS_rfc.loc[:,i]
    rfc_df[i]['VAE'] = VAE_rfc.loc[:,i]
    rfc_df[i] = rfc_df[i].T

trust_df = {}
for i in GDAVIS_trust.columns:
    trust_df[i] = pd.DataFrame()
    trust_df[i]['G-NeuroDAVIS'] = GDAVIS_trust.loc[:,i]
    trust_df[i]['VAE'] = VAE_trust.loc[:,i]
    trust_df[i] = trust_df[i].T
```

# Visualization

```python
for typ in GDAVIS_knn.columns:
    yticks = np.linspace(knn_df[typ].min().min()-0.05, knn_df[typ].max().max()+0.05, 5)[1:-1]
    ylabel = [str(round(i,2)) for i in yticks]
    ylim = [knn_df[typ].min().min()-0.05,knn_df[typ].max().max()+0.05]
    label = knn_df[typ].index
    col = ['#dd8452','#4c72b0']


    path = res_plots+dataname+'/'+dataname+'_'+typ+'_knn.eps'
    CP.SpiderPlot(knn_df[typ], yticks, ylabel, ylim, label, col, path)
```

```python
for typ in GDAVIS_rfc.columns:
    yticks = np.linspace(rfc_df[typ].min().min()-0.05, rfc_df[typ].max().max()+0.05, 5)[1:-1]
    ylabel = [str(round(i,2)) for i in yticks]
    ylim = [rfc_df[typ].min().min()-0.05,rfc_df[typ].max().max()+0.05]
    label = rfc_df[typ].index
    col = ['#dd8452','#4c72b0']


    path = res_plots+dataname+'/'+dataname+'_'+typ+'_rfc.eps'
    CP.SpiderPlot(rfc_df[typ], yticks, ylabel, ylim, label, col, path)
```
