# Importing libraries

```python
import sys
sys.path.append(path)

import os
os.environ['TF_CPP_MIN_LOG_LEVEL'] = '3'

import keras
import ipynbname
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from utils import Datasets as ds 
from utils import NeuroGDAVIS_1 as ngd

dataname = ipynbname.name().rsplit("_")[0]
txt = %pwd
methodname = txt.rsplit("/")[-1]
```

# Importing datasets

```python
X, y = ds.LoadData(dataname, DType='test')
```

```python
n_values = X.shape[0]
input = np.eye(n_values)
```

# Run NeuroGDAVIS

```python
model = ngd.train(input, X, epoch=2000, num_layers=[64, 1024], bs=40)

newpath = res_model+dataname+'/'
if not os.path.exists(newpath):
    os.makedirs(newpath)
# Saving the model
model.encoder.save(newpath+dataname+'_'+methodname+'_encoder')
model.decoder.save(newpath+dataname+'_'+methodname+'_decoder')

# extracting embedding
_, _, x_embedding = model.encoder.predict(input)
newpath = path_emb+dataname+'/'
if not os.path.exists(newpath):
    os.makedirs(newpath)
pd.DataFrame(x_embedding).to_csv(newpath+dataname+'_'+methodname+'.csv')
```
